# Introdução à Teoria das Probabilidadess

Antes de mais nada é preciso saber diferenciar um modelo **determinístico** de um **probabilístico**. Um modelo determinístico é aquele onde dado as condições iniciais, podemos prever a saída de um experimento **com certeza**. Para entendermos o modelo probabilístico e necessário primeiro entender o que é um experimento aleatório. Em um experimento aleatório cada experimento pode ser executado um número infinito de vezes sob condições inalteradas. Por exemplo: quuando lançamos um dado, obtemos uma sequência de valores.
- X = [1, 4, 5, 6, 3, 1, 2, 1, 3, 6, 5, 3, 2, 3, 4, 3, 2, 5, ...]

Não somos capazes de dizer qual será o próximo valor, mas podemos determinar a probabilidade de cada saída possível.

```python
# Import Libraries

import numpy             as np
import pandas            as pd
import matplotlib.pyplot as plt
```

```python
p = 0.6 # probabilidade de sair cara
nsim = 10
nhead = 0
saida = []
for i in range(1, nsim):
    if (np.random.uniform() < p):
        nhead = nhead + 1
        saida.append(1)
    else:
        saida.append(0)

print("Saída:", saida)
print("Probabilidade de sair cara: ", nhead/nsim)
print("Probabilidade de sair coroa: ", 1-nhead/nsim)
```

Desta forma, podemos definir o modelo probabilístico aquele em que em primeiro momento as saídas de cada experimento parecem ser imprevisíveis, mas quando se analisa um grande número de experimentos, um padrão emerge.

Não podemos determinar exatamente o valor da saída de um experimentom, mas as probabilidades de cada saída possível. A distribuição é dada por:

$$
P(X = k) = \frac{n!}{(n-k)!\,k!}\, p^{k}(1-p)^{\,n-k}, \quad k = 0,1,\ldots,n
$$


## Conceitos Fundamentais

- **Espaço Amostral ($\Omega$)**: conjunto de saídas do experimento.
- **Evento (A)**: um elemento do espaço amostral.
- **Relações entre Eventos**:
    - Evento Impossível: $\varnothing$
    - Evento Certo: $\Omega$
    - $A \cup B$: é o evento que ocorre se A **ou** B **ou** ambos ocorrerem.
    - $A \cap B$: é o evento que ocorre se, e somente se, A **e** B ocorrem.
    - $\overline{A}$: é o evento que ocorre se A não ocorre.
    - Evento mutualmente exclusivos: $A \cap B = \varnothing$

Definição frequentista de probabilidade:

```python
p = 0.6
vp = []
nsim_list = []
Nmax = 1000

for n in np.arange(10, Nmax + 1, 10):  # começa em 10, evita divisão por zero
    nhead = 0
    for _ in range(n):
        if np.random.uniform() < p:
            nhead += 1

    vp.append(nhead / n)
    nsim_list.append(n)

plt.figure(figsize=(8, 6))
plt.plot(nsim_list, vp, linewidth=2, label="Valor Simulado", color = 'red')
plt.axhline(y=p, linestyle='--', label="Valor Teórico")
plt.ylabel("Fração de Caras", fontsize=14)
plt.xlabel("Número de Experimentos", fontsize=14)
plt.xlim([0, Nmax])
plt.ylim([0, 1])
plt.legend()
plt.show()
```
